```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = range(0,10)
y = range(1,11)
v = range(10,1000,100)
plt.scatter(x,y,c=v)
plt.show()
```

```python
def gausskernel2D(s=9, pwr=10):        
    myarr = np.linspace(0.2,0.8,s//2)
    myarr = np.concatenate([myarr,np.array([1]),myarr[::-1]])
    myarr *= pwr
    mymat = (np.tile(myarr,(s,1)))
    gknl = np.zeros(mymat.shape)
    for _ in range(10):
        gknl += mymat
        mymat = mymat.T
    return gknl
```

```python
s = 19
mymesh = np.meshgrid(range(0,s),range(0,s))
gknl = gausskernel2D(s, 10)
```

```python
plt.scatter(mymesh[0],mymesh[1],s=gknl,c=gknl,marker='o')
plt.show()
```
